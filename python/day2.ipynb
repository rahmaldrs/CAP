    Create an array with random values
    With a lower limit value (-1);
    Superintendent (100) and;
    The order of the matrix is 5x5.
    Then, determine what are the min, max, mean, median, standard deviation values

```python
rand=np.random.RandomState()
log= rand.randint(-1,100,size=25)
mtx=log.reshape(5,5)
print((mtx),'\nmin=',(np.min(mtx)),'\nmax=',(np.max(mtx)),'\nmean=',(np.mean(mtx)),'\nmedian=',(np.median(mtx)),
     '\nstandard deviation=',(np.std(mtx)))
```
