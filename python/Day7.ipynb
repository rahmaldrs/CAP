2.  Known Ages = [23, 16, 14, 28, 19, 11, 38]
- Find out the age of our youngest Reader
- Find out the oldest reader
- Find out how many years in total our readers have life experience

  Print the message of a, b, c with the way you are 

```python
Known_Ages = [23, 16, 14, 28, 19, 11, 38]
min(Known_Ages)
max(Known_Ages)
len(Known_Ages)
print("youngest Reader", min(Known_Ages))
print("oldest reader",max(Known_Ages))
print("years in total our readers have life experience",sum(Known_Ages))
print("years in total our readers have life experience",max(Known_Ages)-min(Known_Ages))   
```

3.  Every multiple change the value to Hello.

    Every multiple of 7, change the value to Hello World.
    
    If you meet 35 & 70, change the value to Hello !..
    
    And for 100, change the value to End.

```python
a=int(input("a="))
b=int(input("b="))

if(a==7 or b==7):
    print('Hello World.')
elif(a==100 or b==100):
    print('End.')
elif(a*b==35 or a*b==70):
    print('Hello !..')
elif(a*b):
    print('Hello')
```

```python
def number(num):
    
    if num!= 0:
        print("Hello")
        
    if (num%7)==0:
        print("Hello World")
    
    if(num==35):
        print("Hello...")
    elif(num==70):
        print("Hello...")
    if num==100:
        print("End")
    else:
        print("input another number")

input_number=int(input("enter number="))
number(input_number)

while input_number!=100:
    input_number=int(input("enter number="))
    number(input_number)
    
```

5. Buatlah aplikasi kalkulator  

   Kalkulator bisa mengoperasikan +,-,x,/,pangkat


```python
#penjumlahan
def add(a,b):
    return a+b
#pengurangan
def minus(a,b):
    return a-b
#perkalian
def multiply(a,b):
    return a*b
#pembagian
def derivation(a,b):
    return a/b
#pangkat
def exponential(num1, exp):
    return num1**exp
print("Masukkan pilihan operasi kalkulato -\n" \
      "1. Additional\n" \
      "2. Substract \n" \
      "3. Multiply \n" \
      "4. Divide \n" \
      "5. Exponential \n")

select = int(input("Select operations menu form 1, 2, 3, 4, 5: "))

number1 = int(input("Enter first number: "))
number2 = int(input("Enter second number: "))

if select == 1:
    print(number1, "+", number2, "=", add(number1, number2))
elif select == 2:
    print(number1, "-", number2, "=", substract(number1, number2))
elif select == 3:
    print(number1, "x", number2, "=", multiply(number1, number2))
elif select == 4:
    print(number1, "/", number2, "=", divide(number1, number2))
elif select == 5:
    print(number1, "^", number2, "=", exponential(number1, number2))
```
