2.  Known Ages = [23, 16, 14, 28, 19, 11, 38]
- Find out the age of our youngest Reader
- Find out the oldest reader
- Find out how many years in total our readers have life experience

  Print the message of a, b, c with the way you are 

```python
Known_Ages = [23, 16, 14, 28, 19, 11, 38]
min(Known_Ages)
max(Known_Ages)
len(Known_Ages)
print("youngest Reader", min(Known_Ages))
print("oldest reader",max(Known_Ages))
print("years in total our readers have life experience",max(Known_Ages)-min(Known_Ages))   
```

3.  Every multiple change the value to Hello.

    Every multiple of 7, change the value to Hello World.
    
    If you meet 35 & 70, change the value to Hello !..
    
    And for 100, change the value to End.

```python
a=int(input("a="))
b=int(input("b="))

if(a==7 or b==7):
    print('Hello World.')
elif(a==100 or b==100):
    print('End.')
elif(a*b==35 or a*b==70):
    print('Hello !..')
elif(a*b):
    print('Hello')
```
